```python
from matplotlib import pyplot as plt
```

# Линейный график

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("GDP")
plt.ylabel("$");
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bians_sqared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
```

```python
total_error = [x + y for x, y in zip(variance, bians_sqared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='dispercia')
plt.plot(xs, bians_sqared, 'r-.', label='bians sqared')
plt.plot(xs, total_error, 'b:', label='error')
plt.legend(loc=9)
plt.xlabel("Difficult of model")
plt.title("The trade-off between bias and variance");
```

## Группировка значений столбчатых графиков

```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
```

```python
plt.bar([x + 5 for x in histogram.keys()], # Сдвигаем на 5 единиц столбец
       histogram.values(), # Берем значения для высоты столбцов
       10, # Ширина столбца
       edgecolor=(0, 0, 0)) # Черная окантовка для каждого столбца
plt.axis([-5, 105, 0, 5]) # Назначение размерности осей
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("Number of Students")
plt.title("Distribution of ratings");
```

# Диаграммы рассеяния

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy = (friend_count, minute_count), # Задать метку
                 xytext = (5, -5), # Сместить ее
                 textcoords = 'offset points')

plt.title("Number minutes vs number friends")
plt.xlabel("Number friends")
plt.ylabel("Number minuter in site");
```
