```python
Vector = List[float]
```

```python
def add(v: Vector, w: Vector) -> Vector:
    assert len(v) == len(w)
    return [v_i + w_i for v_i, w_i in zip(v, w)]
```

```python
print(add([1, 2, 3], [4, 5, 6]))
```

```python
def subtract(v: Vector, w: Vector) -> Vector:
    assert len(v) == len(w)
    return [v_i - w_i for v_i, w_i in zip(v, w)]
```

```python
print(subtract([5, 7, 9], [4, 5, 6]))
```

```python
def vector_sum(vectors: List[Vector]) -> Vector:
    assert vectors
    
    num_elements = len(vectors[0])
    assert all(len(v) == num_elements for v in vectors)
    
    return [sum(vector[i] for vector in vectors) for i in range(num_elements)]
```

```python
def scalar_multiply(c: float, v: Vector) -> Vector:
    return [c * v_i for v_i in v]
```

```python
print(scalar_multiply(2, [1, 2, 3]))
```

```python
def vector_mean(vectors: List[Vector]) -> Vector:
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum(vectors))
```

```python
print(vector_mean([[1, 2], [3, 4], [5, 6]]))
```

```python
def dot(v: Vector, w: Vector) -> float:
    """Вычисляем v_1 * w_1 + ... + v_n * w_n"""
    assert len(v) == len(w)
    return sum(v_i * w_i for v_i, w_i in zip(v, w))
```

```python
print(dot([1, 2, 3], [4, 5, 6]))
```

```python
def sum_of_squares(v: Vector) -> float:
    return dot(v, v)
```

```python
print(sum_of_squares([1, 2, 3]))
```

```python
def magitude(v: Vector) -> float:
    return math.sqrt(sum_of_squares(v))
```

```python
print(magitude([3, 4]))
```

```python
def distance(v: Vector, w: Vector) -> float:
    return magitude(subtract(v, w))
```

```python
print(distance([0, 0, 0], [0, 0, 1]))
```

## Матрицы

```python
Matrix = List[List[float]]
```

```python
A = [[1, 2, 3],
     [4, 5, 6]] # Матрица A с 2 строками и 3 столбцами

B = [[1, 2],
     [3, 4],
     [5, 6]] # Матрица B с 3 строками и 2 столбцами
```

```python
def shape(A: Matrix) -> Tuple[int, int]:
    """Возвращает число строк A, число стобцов A"""
    
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0 # Число элементов в первой строке
    
    return num_rows, num_cols
```

```python
print(shape(A))
```

```python
def get_row(A: Matrix, i: int) -> Vector:
    """Возвращает i-ю строку A, как вектор"""
    return A[i]
```

```python
def get_column(A: Matrix, j: int) -> Vector:
    """Возвращает j-й столбец A, как вектор"""
    return [A_i[j] for A_i in A]
```

```python
def make_matrix(num_rows: int,
                num_cols: int,
                entry_fn: Callable[[int, int], float]) -> Matrix:
    """Возвращает матрицу размера num_rows x num_cols, 
    чей (i, j)-й элемент является функцией entry_fn(i, j)"""
    return [[entry_fn(i, j)             # Создает список с учетом i
             for j in range(num_cols)]  # [entry_fn(i, 0), ... ]
             for i in range(num_rows)]  # Создать один список для каждого i
```

```python
def identity_matrix(n: int) -> Matrix:
    """Создает единичную квадратную матрицу"""
    return make_matrix(n, n, lambda i, j: 1 if i == j else 0)
```

```python
print(identity_matrix(5))
```

```python
def mean(xs: List[float]) -> float:
    return sum(xs) / len(xs)
```

```python
# Символы подчеркивания указывают на то, что эти функции являются приватными
# тк они предназначены для вызова 
# из нашей функции median, а не другими людьми
# использующими нашу статистическую библиотеку
```

```python
def _median_odd(xs: List[float]) -> float:
    """Если len(xs) является нечетной, то медиана - это срединный элемент"""
    return sorted(xs)[len(xs) // 2]
```

```python
def _median_even(xs: List[float]) -> float:
    """Если len(xs) является четной, то она является средним значением двух срединных элементов"""
    sorted_xs = sorted(xs)
    hi_midpoint = len(xs) // 2 
    return (sorted_xs[hi_midpoint - 1] + sorted_xs[hi_midpoint]) / 2
```

```python
def median(v: List[float]) -> float:
    return _median_even(v) if len(v) % 2 == 0 else _median_odd(v)
```

```python
print(median([1, 10, 2, 9, 5])) # 5
print(median([1, 9, 2, 10])) # (2 + 9) / 2 = 5.5
```

```python
print(median(num_friends)) # 6
```

```python
def quantile(xs: List[float], p: float) -> float:
    """Возвращает значение p-го процентиля в x"""
    p_index = int(p * len(xs)) # Преобразует % в индекс списка
    return sorted(xs)[p_index]
```

```python
print(quantile(num_friends, 0.10)) # 1
print(quantile(num_friends, 0.25)) # 3
print(quantile(num_friends, 0.75)) # 9
print(quantile(num_friends, 0.90)) # 13
```

```python
def mode(x: List[float]) -> List[float]:
    """Возвращает список, т.е. может быть больше 1 моды"""
    counts = Counter(x)
    max_count = max(counts.values())
    return [x_i for x_i, count in counts.items()
            if count == max_count]
```

```python
print(set(mode(num_friends))) # {1, 6}
```

```python
# Ключевое слово "range" (размах) в Python уже имеет
# свой смысл, поэтому выберем другое
```

```python
def data_range(xs: List[float]) -> float:
    return max(xs) - min(xs) # Размах равен 0 когда max и min эквивалентны
```

```python
print(data_range(num_friends)) # 99
```

```python
def de_mean(xs: List[float]) -> List[float]:
    """Транслировать xs путем вычитания его среднего (результат имеет нулевое среднее)"""
    x_bar = mean(xs)
    return [x - x_bar for x in xs]
```

```python
def variance(xs: List[float]) -> float:
    """Почти среднеквадратическое отклонение от среднего"""
    assert len(xs) >= 2 # Дисперсия требует наличия не менее двух элементов
    
    n = len(xs)
    deviations = de_mean(xs)
    return sum_of_squares(deviations) / (n - 1)
```

```python
print(variance(num_friends)) # 81.54 < x < 81.55
```

```python
def standard_deviation(xs: List[float]) -> float:
    """Стандартное отклонение - корень из дисперсии"""
    return math.sqrt(variance(xs))
```

```python
standart_deviation(num_friends) # 9.02 < x < 9.043
```

```python
# При выбросах этот подход не практичен, т.к. у одного человека максимально может быть 200 друзей
# тогда бы значение уыеличилось на 60% и делать выводы было бы неверно
# по этому нужно считать интерквартальный размах (для 75% данных и для оставшихся 25%)
```

```python
def interquartile_range(xs: List[float]) -> float:
    """Возвращает разницу между 75% и 25% квартилями"""
    return quantile(xs, 0.75) - quantile(xs, 0.25)
```

```python
interquartile_range(num_friends) # 6
```

```python
def covariance(xs: List[float], ys: List[float]) -> float:
    assert len(xs) == len(ys)
    return dot(de_mean(xs), de_mean(ys)) / (len(xs) - 1)
```

```python
covariance(num_friends, daily_hours) # < 22.43 / 60
```

```python
def correlation(xs: List[float], ys: List[float]) -> float:
    """Измеряет степень, с которой xs и ys  варьируются в тандеме вокруг своих средних"""
    stdev_x = standard_deviation(xs)
    stdev_y = standard_deviation(ys)
    if stdev_x > 0 and stdev_y > 0:
        return covariance(xs, ys) / stdev_x / stdev_y
    else:
        return 0    # если вариации нет, то корреляция равна 0
```

```python
correlation(num_friends, daily_minutes) # < 0.25
```

```python
correlation(num_friends, daily_hours) # < 0.25
```

# Вероятность 

```python
def uniform_cdf(x: float) -> float:
    """Возвращает вероятность, что равномерно распределенная случайная величина <= x"""
    if x < 0:   return 0    # равномерная величина никогда не бывает меньше 0
    elif x < 1: return x    # Например, P(x <= 0.4) = 0.4
    else:       return 1    # равномерная величина всегда меньше 1
```

```python
SQRT_TWO_PI = math.sqrt(2 * math.pi)

def normal_pdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (SQRT_TWO_PI * sigma))
```

```python
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs,[normal_pdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
plt.plot(xs,[normal_pdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
plt.plot(xs,[normal_pdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
plt.plot(xs,[normal_pdf(x,mu=-1)   for x in xs],'-.',label='mu=-1,sigma=1')
plt.legend()
plt.title("Нормальные функции плотности вероятности");
```

```python
def normal_cdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2
```

```python
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs,[normal_cdf(x,sigma=1) for x in xs],'-',label='mu=0,sigma=1')
plt.plot(xs,[normal_cdf(x,sigma=2) for x in xs],'--',label='mu=0,sigma=2')
plt.plot(xs,[normal_cdf(x,sigma=0.5) for x in xs],':',label='mu=0,sigma=0.5')
plt.plot(xs,[normal_cdf(x,mu=-1) for x in xs],'-.',label='mu=-1,sigma=1')
plt.legend(loc=4) # внизу справа
plt.title("Нормальные коммулятивные функции распределения");
```
